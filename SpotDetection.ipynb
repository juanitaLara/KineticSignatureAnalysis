```python
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from skimage.color import rgb2gray

#Open image that will act as mask in order to find the ROIs 
image = data.imread('Labview_sr_image.tif')

fig = figure(figsize=(15,15))
imshow(image, cmap = cm.Greys_r)
colorbar()

print len(image)
```

```python
image_gray = rgb2gray(image)
```

```python
#Morphological filtering
#Opening: erosion -> dilation. Light regions that are smaller than the structuring element are removed. 
#Result: The dilation operation ensures that light regions that are larger than the structuring element retain 
#their original size. 
#Closing: dilation -> erosion. Dark regions that are smaller than the structuring element are removed. 
#Result: The dilation operation that dark regions that are larger than the structuring element retain their original size. 

from skimage.morphology import erosion, dilation, opening, closing
from skimage.morphology import convex_hull_image
from skimage.morphology import disk
from scipy import ndimage

#Zoom in to a region in the center
sample_im = image_gray[1000:1200,1000:1200]

#Set the diamter of the structuring element (disk)
selem = disk(1)

#Remove small white regions
open_im = opening(sample_im, selem)

#Remove small black hole
closed_im = closing(open_im, selem)

#Reconstructed image
reconstruction = ndimage.binary_propagation(closed_im, mask=sample_im)

#Dilated reconstruction
dilated_rec = dilation(dilation(reconstruction,selem),selem)

fig = figure(figsize=(60, 15))

ax = fig.add_subplot(141)
ax.imshow(sample_im, cmap=plt.cm.gray)
ax.set_title('Original', fontsize=30)
ax.axis('off')

ax = fig.add_subplot(142)
ax.imshow(open_im, cmap=plt.cm.gray)
ax.set_title('Opened', fontsize=30)
ax.axis('off')

ax = fig.add_subplot(143)
ax.imshow(closed_im, cmap=plt.cm.gray)
ax.set_title('Closed', fontsize=30)
ax.axis('off')

ax = fig.add_subplot(144)
ax.imshow(dilated_rec, cmap=plt.cm.gray)
ax.set_title('Dilated Reconstruction', fontsize=30)
ax.axis('off')

show()
```

```python
#Blobs are found using the Determinant of Hessian method
blobs_doh = blob_doh(dilated_rec, max_sigma=31, threshold=0.02)
```

```python
#Blobs are found using the Laplacian of Gaussian (LoG) method
blobs_log = blob_log(dilated_rec, max_sigma=30, threshold=.1)
```

```python
# Compute radii in the 3rd column.
blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)

blobs_list = [blobs_log, blobs_doh]
colors = ['yellow', 'lime']
titles = ['Laplacian of Gaussian','Determinant of Hessian']
sequence = zip(blobs_list, colors, titles)

for blobs, color, title in sequence:
    fig, ax = plt.subplots(1, 1, figsize = (10,10))
    ax.set_title(title)
    ax.imshow(dilated_rec, cmap = cm.Greys_r)
    for blob in blobs:
        y, x, r = blob
        r = plt.Rectangle((x-r, y-r), 2*r, 2*r, color=color, linewidth=1, fill=False)
        
        ax.add_patch(r)

plt.show()
fig.savefig('Co-PAINT_5T_P1Cy3b_5nM001.trace.mat.blob.eps')
```
